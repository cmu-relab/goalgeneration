## Analyze Symmetries using Human Labels

This notebook conducts a series of experiments to survey different ways to direct refinement relationships to reduce inaccuracy, including inacurracies due to symmetric relations.

```python
import csv

data_path = 'data1'

rows = [[r[0], r[1], r[3]] for r in csv.reader(open('%s/comps_final.csv' % data_path, 'r'))]
```

```python
from openai import OpenAI

client = OpenAI()

def prompt_model(prompt):
    response = client.chat.completions.create(
      model="gpt-4o-2024-08-06",
      #model='o3-mini-2025-01-31',
      #model='gpt-4o-mini',
      messages=[
        {
          "role": "system",
          "content": "You are a helpful assistant."
        },
        {
          "role": "user",
          "content": prompt
        }
      ]
    )
    return response.choices[0].message.content
```

```python
# 

prompt1 = """Read the following statement then answer the question by responding True or False.

Statement: %s

Question: The statement describes how to %s.
Answer: """

prompt2 = """Read the following statement then answer the question by responding True or False.

Statement: %s

Question: The statement describes why we need to %s.
Answer: """

print('Comps 1 ', end='')
comps1 = []
for row in rows:
    p1 = prompt1 % (row[0], row[1])
    p2 = prompt1 % (row[1], row[0])
    p3 = prompt2 % (row[0], row[1])
    p4 = prompt2 % (row[1], row[0])
        
    a = []
    for i in range(10):
        r1 = prompt_model(p1)
        r2 = prompt_model(p2)
        r3 = prompt_model(p3)
        r4 = prompt_model(p4)
        a.append([r1[0], r2[0], r3[0], r4[0]])

    comps1.append([row[0], row[1], a])
    print(' .', end='')
print()
```

```python
prompt3 = """Compare the following two statements and choose the best answer. Do not comment or elaborate.

S1: %s
S2: %s

A) Statement S1 describes a software operation more than Statement S2
B) Statement S2 describes a software operation more than Statement S1
C) Statement S1 means more or less the same thing as statement S2
Answer: """

prompt4 = """Compare the following two statements and choose the best answer. Do not comment or elaborate.

S1: %s
S2: %s

A) Statement S1 describes a general goal more than Statement S2
B) Statement S2 describes a general goal more than Statement S1
C) Statement S1 means more or less the same thing as statement S2
Answer: """


prompt5 = """Compare the following two statements and choose the best answer. Do not comment or elaborate.

S1: %s
S2: %s

A) If statement S1 is true, then later statement S2 will become true
B) If statement S2 is true, then later statement S2 will become true
C) If there is no order between S1 or S2 about when S1 or S2 is true
Answer: """

print('Comps 2 ', end='')
comps2 = []
for row in rows:
    p3a = prompt3 % (row[0], row[1])
    p3b = prompt3 % (row[1], row[0])
    p4a = prompt4 % (row[0], row[1])
    p4b = prompt4 % (row[1], row[0])
    p5a = prompt5 % (row[0], row[1])
    p5b = prompt5 % (row[1], row[0])
    
    a = []
    for i in range(5):
        r3a = prompt_model(p3a)
        r3b = prompt_model(p3b)
        r4a = prompt_model(p4a)
        r4b = prompt_model(p4b)
        r5a = prompt_model(p5a)
        r5b = prompt_model(p5b)
        a.append([r3a[0], r3b[0], r4a[0], r4b[0], r5a[0], r5b[0]])

    comps2.append([row[0], row[1], a])
    print(' .', end='')
print()
```

```python
for [x, y, a] in comps2:
    print('S1: %s' % x)
    print('S2: %s' % y)
    print('Ans: %s' % a)
    print()
```

```python
prompt6 = """Read the following statement and choose the best category for the statement. Do not comment or elaborate.

Statement: %s

A) The statement describes a goal that must be clarified, refined or disambiguated before it can be implemented by a developer
B) The statement describes a software feature that a developer can implement 
Answer: """

prompt5 = """Read the following statement and choose the best category for the statement. Do not comment or elaborate.

Statement: %s

A) The statement describes an action that is a person's goal about what they want to do.
B) The statement describes an action performed by a function in an application that can be tested for success or failure. 
Answer: """

print('Comps 3 ', end='')
comps3 = []
for row in rows:
    px = prompt5 % row[0]
    py = prompt5 % row[1]
    
    a = []
    for i in range(10):
        rx = prompt_model(px)
        a.append(rx[0])

    b = []
    for i in range(10):
        ry = prompt_model(py)
        b.append(ry[0]) 

    comps3.append([row[0], a])
    comps3.append([row[1], b])
    print(' .', end='')
print()
```

```python
from collections import Counter

def get_majority(answers):
    c = Counter()
    for a in answers:
        c[a] += 1
    m = c.most_common(2)
    if len(m) == 1:
        return [m[0][0], m[0][1]]
    elif m[0][1] == m[1][1]:
        return [m[0][0], m[1][0], m[0][1]]
    else:
        return [m[0][0], m[0][1]]

for i in range(0, len(comps3), 2):
    print('S1: %s' % comps3[i][0])
    print('Ans: %s >> %s' % (comps3[i][1], get_majority(comps3[i][1])))
    print('S2: %s' % comps3[i+1][0])
    print('Ans: %s >> %s' % (comps3[i+1][1], get_majority(comps3[i+1][1])))
    print()
```
